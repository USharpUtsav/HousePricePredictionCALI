```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
df=pd.read_csv('housing.csv')
```

```python
df
```

**WE WONT USE OCEAN PROXIMITY AS OUR PROJECT DO NOT REQUIRES SUCH:**

```python
df=df.drop(columns='ocean_proximity')
```

```python
df.shape
```

```python
df.info()
```

**DEALING WITH NULL/NAN VALUES** 

```python
# Drop rows with NaN values
df_cleaned = df.dropna()

# Check the info of the cleaned dataframe
df_cleaned.info()
```

```python
df_cleaned.describe()
#A large apartment complex with 200 apartments, each having 3 bedrooms, could easily contribute to a high total like 6445 bedrooms. 
#In this case, the number of units (apartment blocks) would need to be quite large to reach that figure.commercial apartments
```

**Correlation between features of the dataset**


```python
# Rename the column 'median_house_value' to 'Price'
df_cleaned = df_cleaned.rename(columns={'median_house_value': 'price'})
df_cleaned = df_cleaned.rename(columns={'housing_median_age': 'house_age'})
df_cleaned
```

```python
from numpy.ma.core import size

cor= df_cleaned.corr()

#check correlation using a heatmap
plt.figure(figsize=(11,11))
sns.heatmap(cor,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':9},cmap='Greens')


#MEDIAN INCOME COULD BE A GOOD PREDICTOR FOR OUR MODEL AS IT HAS THE HIGHEST CORRELATION WITH PRICE
```

```python
df_cleaned.hist(figsize=(20,12))
```

**DEALING WITH SKEWNESS IN DATA**

```python
df_cleaned['total_bedrooms']=np.log(df_cleaned['total_bedrooms']+1)
df_cleaned['population']=np.log(df_cleaned['population']+1)
```

```python
df_cleaned['total_rooms']=np.log(df_cleaned['total_rooms']+1)
df_cleaned['households']=np.log(df_cleaned['households']+1)
```

```python
df_cleaned.hist(figsize=(20,12))
```

**Splitting the data and Target**


```python
X = df_cleaned.drop(['price'], axis=1)
Y = df_cleaned['price']
```

```python
X
```

```python
Y
```

```python
#Splitting the data into Training data and Test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Standardize 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the model
lr_model = LinearRegression()

# Fit the model on the training data
lr_model.fit(X_train_scaled, Y_train)

# Make predictions on the test set
Y_pred = lr_model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Linear Regression - MSE: {mse}")
print(f"Linear Regression - R-squared: {r2}")
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_scaled, Y_train)

Y_pred_rf = rf_model.predict(X_test_scaled)

rf_mse = mean_squared_error(Y_test, Y_pred_rf)
rf_r2 = r2_score(Y_test, Y_pred_rf)

print(f"Random Forest - MSE: {rf_mse}")
print(f"Random Forest - R-squared: {rf_r2}")
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_model.fit(X_train_scaled, Y_train)

Y_pred_gb = gb_model.predict(X_test_scaled)

gb_mse = mean_squared_error(Y_test, Y_pred_gb)
gb_r2 = r2_score(Y_test, Y_pred_gb)

print(f"Gradient Boosting - MSE: {gb_mse}")
print(f"Gradient Boosting - R-squared: {gb_r2}")
```

```python
import xgboost as xgb

xgb_model = xgb.XGBRegressor(n_estimators=100, random_state=42)
xgb_model.fit(X_train_scaled, Y_train)

Y_pred_xgb = xgb_model.predict(X_test_scaled)

xgb_mse = mean_squared_error(Y_test, Y_pred_xgb)
xgb_r2 = r2_score(Y_test, Y_pred_xgb)

print(f"XGBoost - MSE: {xgb_mse}")
print(f"XGBoost - R-squared: {xgb_r2}")
```

**HYPERPARAMETER TUNNING**

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import pickle

# Define the XGBRegressor model
xgb_model = xgb.XGBRegressor(n_estimators=100, random_state=42)

# Set the parameter grid to search over
param_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0],
    'gamma': [0, 0.1, 0.2],
    'reg_alpha': [0, 0.1, 0.5],
    'reg_lambda': [0.5, 1, 1.5]
}

# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)

# Fit the grid search to the data
grid_search.fit(X_train_scaled, Y_train)

# Get the best parameters from the grid search
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Train the model with the best parameters
best_xgb_model = grid_search.best_estimator_

# Make predictions
Y_pred_xgb = best_xgb_model.predict(X_test_scaled)

# Evaluate the model
xgb_mse = mean_squared_error(Y_test, Y_pred_xgb)
xgb_r2 = r2_score(Y_test, Y_pred_xgb)

print(f"XGBoost - MSE: {xgb_mse}")
print(f"XGBoost - R-squared: {xgb_r2}")
```

```python
import pickle

# Save the best XGBoost model using pickle
with open('best_xgb_model.pkl', 'wb') as f:
    pickle.dump(best_xgb_model, f)

# Save the scaler (to use for scaling user input in the Streamlit app)
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

print("Model and scaler saved successfully.")
```
