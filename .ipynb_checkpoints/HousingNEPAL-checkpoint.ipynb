```python
import re

def convert_to_sqft(value):
    value = str(value).strip().lower()  # Ensure the value is a string and clean it
    
    # Regular expression to extract the first valid floating-point or integer number
    match = re.search(r'(\d+\.\d+|\d+)', value)
    
    if not match:
        return None  # or you could set a default, like 0

    try:
        number = float(match.group(1))  # Extract the numeric part as a float
        
        if 'aana' in value:
            return number * 342
        elif 'kattha' in value:
            return number * 3388
        elif 'sq. mtr' in value or 'mtr' in value:
            return number * 10.764
        elif 'sq. ft' in value or 'ft' in value:
            return number  # Already in sq. ft
        else:
            # Assume the value is in 'aana' if no unit is specified
            return number * 342
    except ValueError:
        return None  # or set a default, like 0, if conversion fails

# Apply conversion to the 'LAND AREA' column
df['LAND AREA (sq. ft)'] = df['LAND AREA'].apply(convert_to_sqft)

# Display the dataframe with the new column
print(df[['LAND AREA', 'LAND AREA (sq. ft)']].head())
```

```python
# Apply the conversion function to 'LAND AREA' column and create a new 'LAND AREA (sq. ft)' column
df['LAND AREA (sq. ft)'] = df['LAND AREA'].apply(convert_to_sqft)

# Drop the original 'LAND AREA' column if you no longer need it
df.drop('LAND AREA', axis=1, inplace=True)

# Optionally, you can rename the new column to 'LAND AREA' for consistency
df.rename(columns={'LAND AREA (sq. ft)': 'LAND AREA'}, inplace=True)
```

```python
import pandas as pd

# Function to handle parking values and convert them into total number of vehicles
def convert_parking_to_vehicles(parking_value):
    cars = 0
    bikes = 0
    
    # Handle cars: check for both "CaRs" and possible ranges like "10-15"
    if "CaRs" in parking_value:
        cars_part = parking_value.split('CaRs')[0].strip()  # Extract part before 'CaRs'
        if '-' in cars_part:  # If there's a range, take the average
            car_range = cars_part.split('-')
            cars = (int(car_range[0]) + int(car_range[1])) / 2  # Averaging the range
        else:
            cars = int(cars_part)  # Convert directly to int if no range
    
    # Handle bikes: check for both "Bikes" and possible ranges like "3-5"
    if "Bikes" in parking_value:
        bikes_part = parking_value.split('Bikes')[0].split('&')[-1].strip()  # Extract part before 'Bikes'
        if bikes_part:  # If bikes part is not empty, convert it to integer
            if '-' in bikes_part:  # If there's a range, take the average
                bike_range = bikes_part.split('-')
                bikes = (int(bike_range[0]) + int(bike_range[1])) / 2  # Averaging the range
            else:
                bikes = int(bikes_part)  # Convert directly to int if no range
    # If no bikes are specified, bikes remain 0

    # Return the sum of cars and bikes as a float
    return float(cars + bikes)

# Apply the conversion to create the new column
df['PARKING (No of Vehicle)'] = df['PARKING'].apply(lambda x: convert_parking_to_vehicles(str(x)))

# Check the result
print(df[['PARKING', 'PARKING (No of Vehicle)']].head())
```

# DEPENDENT VARIABLE(PRICE) standardized

```python
# To show the unique values in the 'PRICE' column
print(df['PRICE'].unique())
```

```python
import numpy as np

def standardize_price(price):
    # If the price is already a float, just return it (no need to process)
    if isinstance(price, float):
        return price
    
    # Clean the price string
    price = price.replace('Rs. ', '').replace(',', '').strip()  # Remove 'Rs.' and commas

    print(f"After cleaning: {price}")  # Debug print
    
    # Check for known units and convert accordingly
    if 'Cr' in price:
        # Remove 'Cr' and convert to numerical value in crores
        price = price.replace('Cr', '').strip()
        try:
            price = float(price) * 10**7  # Convert crores to rupees
        except ValueError:
            print(f"Warning: Could not convert '{price}' to float after removing 'Cr'. Returning NaN.")  # Debug print
            return np.nan
    elif 'Lac' in price:
        # Remove 'Lac' and convert to numerical value in lakhs
        price = price.replace('Lac', '').strip()
        try:
            price = float(price) * 10**5  # Convert lakhs to rupees
        except ValueError:
            print(f"Warning: Could not convert '{price}' to float after removing 'Lac'. Returning NaN.")  # Debug print
            return np.nan
    elif '/m' in price:
        # Remove '/m' and convert to numerical value assuming monthly rent
        price = price.replace('/m', '').strip()
        try:
            price = float(price) * 12  # Annualize the monthly value
        except ValueError:
            print(f"Warning: Could not convert '{price}' to float after removing '/m'. Returning NaN.")  # Debug print
            return np.nan
    elif 'Price on call' in price:
        return np.nan  # Represent unknown prices as NaN
    else:
        # Try to convert directly to float
        try:
            price = float(price)  # Convert to float directly
        except ValueError:
            print(f"Warning: Could not convert '{price}' to float. Returning NaN.")  # Debug print
            return np.nan  # If conversion fails, return NaN

    # Return the price as a float (this will ensure the value is in rupees)
    print(f"Final price after conversion: {price}")  # Debug print
    return float(price)

# Apply the standardization function to the 'PRICE' column
df['PRICE'] = df['PRICE'].apply(standardize_price)

# Check the updated DataFrame
print(df['PRICE'].unique())
```

**CREATING A NEW COLOUMN TO FIND OUT THE AGE OF THE PROPERTY**

```python
# Function to convert Nepali year to Gregorian year
def nepali_to_gregorian(nepali_year):
    # Remove the 'B.S' part and convert to an integer
    nepali_year = int(nepali_year.replace(' B.S', '').strip())
    # Subtract 56.7 years to get the approximate Gregorian year
    gregorian_year = nepali_year - 57
    return gregorian_year

# Apply the conversion to the 'BUILT YEAR' column
df['BUILD_YEAR_GREGORIAN'] = df['BUILT YEAR'].apply(nepali_to_gregorian)

# Calculate the age of the property
current_year = 2024
df['AGE'] = current_year - df['BUILD_YEAR_GREGORIAN']

# Drop the 'BUILT YEAR' column as it is no longer needed
df = df.drop('BUILT YEAR', axis=1)

# Check the updated DataFrame
print(df.head())
```

```python
# Convert 'AGE' column to float
df['AGE'] = df['AGE'].astype(float)

# Convert 'AGE' column to float
df['BUILD_YEAR_GREGORIAN'] = df['BUILD_YEAR_GREGORIAN'].astype(float)

# Check the updated DataFrame
print(df.dtypes)
```

# MODELLING TRAIN/TEST SPLIT:-

```python
from sklearn.model_selection import train_test_split

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import cross_val_score

# Define the models to evaluate
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
    'SVR': SVR(),
    'KNN': KNeighborsRegressor(n_neighbors=5)
}
```

```python
# Perform cross-validation for each model
results = {}
for model_name, model in models.items():
    # Perform 5-fold cross-validation using the scaled training data
    cv_results = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='r2')  # 'r2' for R-squared
    results[model_name] = cv_results

    # Print the results for each model
    print(f"{model_name} - Cross-Validation R-squared Scores: {cv_results}")
    print(f"{model_name} - Mean R-squared: {cv_results.mean()} ± {cv_results.std()}")
    print("="*50)

# Store the results in a DataFrame for easier comparison
cv_results_df = pd.DataFrame(results)

# Display the results
print("\nCross-Validation Results (Mean R-squared and Standard Deviation for each model):")
print(cv_results_df)

# Plot the results
plt.figure(figsize=(10, 6))
sns.boxplot(data=cv_results_df, palette='viridis')
plt.title('Cross-Validation R-squared Scores for Different Models')
plt.ylabel('R-squared')
plt.show()
```

**HYPERPARAMETER TUNNING AND CHOOSING THE BEST MODEL FROM RANDOM FOREST AND GRADIENT BOOSTING SINCE THEY ARE PERFORMING WELL**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR

# Define the models
rf = RandomForestRegressor(random_state=42)
gb = GradientBoostingRegressor(random_state=42)

# Define the parameter grids for Random Forest and Gradient Boosting
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

gb_param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}

# Initialize GridSearchCV for Random Forest
rf_grid_search = GridSearchCV(estimator=rf, param_grid=rf_param_grid, cv=5, n_jobs=-1, scoring='r2')

# Initialize GridSearchCV for Gradient Boosting
gb_grid_search = GridSearchCV(estimator=gb, param_grid=gb_param_grid, cv=5, n_jobs=-1, scoring='r2')

# Fit the models
rf_grid_search.fit(X_train_scaled, y_train)
gb_grid_search.fit(X_train_scaled, y_train)

# Get the best parameters and score for each model
rf_best_params = rf_grid_search.best_params_
rf_best_score = rf_grid_search.best_score_

gb_best_params = gb_grid_search.best_params_
gb_best_score = gb_grid_search.best_score_

# Print the results
print(f"Best Random Forest Parameters: {rf_best_params}")
print(f"Best Random Forest Cross-Validation R-squared Score: {rf_best_score:.4f}")

print(f"Best Gradient Boosting Parameters: {gb_best_params}")
print(f"Best Gradient Boosting Cross-Validation R-squared Score: {gb_best_score:.4f}")

# Evaluate on test set
rf_best_model = rf_grid_search.best_estimator_
gb_best_model = gb_grid_search.best_estimator_

rf_test_score = rf_best_model.score(X_test_scaled, y_test)
gb_test_score = gb_best_model.score(X_test_scaled, y_test)

print(f"Random Forest Test R-squared: {rf_test_score:.4f}")
print(f"Gradient Boosting Test R-squared: {gb_test_score:.4f}")
```
